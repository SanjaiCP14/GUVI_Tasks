```python
pas=""
def checkpassword(str1):
  a=0
  b=0
  c=0
  d=0
  for i in str1:
    if(i.isdigit()):
      a+=1
    elif(i.isupper()):
      b+=1
    elif(i.islower()):
      c+=1
    elif(((i.isalnum()) or (i.isdecimal()) or (i.isdigit())) == False):
      d+=1
  if(a>=1 and b>=1 and c>=1 and d>=1):
    return True
  else:
    return False

def checkusername(str2):
  index1 = str2.rfind("@")
  index2 = str2.rfind(".")
  if((index1<index2) and (index1!=-1) and (index2!=-1)):
      if((index1!=index2-1) and (index1!=0)):
          if(str2[0].isalpha()):
                return True
          else:
                return False
      else:
            return False
  else:
        return False

def forgot_method():
  global pas
  print("Your Password is" + pas)

def check_password():
  global pas
  count=0
  while count==0:
      password = input("Password: ")
      if(password==pas):
        print("Logged In successfully")
        count=1
      else:
        print("Enter correct password")
        forget = input("Forget Password(Yes/No): ")
        if(forget=="Yes"):
          forgot_method()

def login_method():
  global pas
  print("Login")
  username = input("Username: ")
  count=0
  try:
      f = open("credentials.txt","r")
      text = f.readlines()
      cred=[]
      f.close()
      for i in text:
          j=i.replace("\n","")
          cred = j.strip().split("-")
          if(username==cred[0]):
              count=1
              pas=cred[1]
              break
  except:
      pass 
  
  if(count==1):
    check_password()

  else:
    print("Not Registered Yet. Please register")
    register_method()

def register_method():
  i=0
  print("Register")
  while i==0:
    username = input("UserName: ")
    if(checkusername(username)):
      i=1
      break
    else:
      print("Enter Correct username")
    
  while i==1:
    password = input("Password: ")
    if(5<len(password)<16):
      if(checkpassword(password)):
        i=2
        break
      else:
        print("Enter Valid password")
    else:
      print("Enter Valid password")
  
  if(i==2):
    f=open("credentials.txt","a")
    f.write(username + "-" + password +"\n")
    f.close()
    print("Successfully registered")
    login_method()


login = input("Do you want to Login (Yes/No): ")
if(login=="Yes"):
  login_method()
elif(login=="No"):
  register = input("Do you want to Register(Yes/No): ")
  if(register=="Yes"):
    register_method()
```
